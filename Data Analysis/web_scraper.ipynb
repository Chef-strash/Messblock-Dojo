```python
def extract(link, data):
    response= requests.get(link)
    soup= BeautifulSoup(response.text, 'html.parser')


    dict={'Quote':{'quote':[], 'tags':[]}, 'author':[], 'about_author':[]}
    for sp in soup.find_all('div', class_='quote'):
        dict['Quote']['quote'].append(sp.find('span', class_='text').text[1:-1])
        dict['author'].append(sp.find('small').text)
        dict['about_author'].append(sp.find('a').get('href'))
        tags=[]
        for tag in sp.find_all('a', class_='tag'):
            tags.append(tag.text)
        dict['Quote']['tags'].append(tags)

    for i in range(len(dict['Quote']['quote'])):
        tags=''
        for tag in (dict['Quote']['tags']):
            tags=','.join(tag)
        data.append([dict['Quote']['quote'][i], dict['author'][i], tags, dict['about_author'][i]])
    
    return data
```

```python
link_p="https://quotes.toscrape.com/page/"

for i in range(2,11):
    link= link_p + str(i) + '/'
    data= extract(link, data)
    
```

```python
df= pd.DataFrame(data, columns=['Quotes', 'Author', 'Tags', 'About'])
df.to_csv('Quotes.csv')
```

**Extract author details**

```python
def extract_author_det(link, data):
    response= requests.get(link)
    soup= BeautifulSoup(response.text, 'html.parser')

    name= soup.find('h3', class_='author-title').text
    DOB_= soup.find('span', class_='author-born-date').text
    place= soup.find('span', class_='author-born-location').text[3:-1]
    det= soup.find('div', class_='author-description').text[1:-1]
    det= det.lstrip()[10:-1]

    data.append([name, DOB_, place, det])

    return data
```

```python
link='https://quotes.toscrape.com/author/J-K-Rowling/'

d= []
d= extract_author_det(link, d)
print(d)
```

**Taking input as a name and scrapoing its wikipedia**

```python
title=str(input('Enter your search: '))
title=title.replace(' ','+')

link= 'https://www.google.com/search?q=' + title + '+wikipedia'
```

```python
print(link)
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

res = requests.get(link, headers=headers)
soup = BeautifulSoup(res.text,'html.parser')
```

```python
# Find all <a> tags and look for Wikipedia links
wikipedia_link = None
for a in soup.find_all('a'):
    href = a.get('href')
    if href and 'en.wikipedia.org' in href:
        # Google search links are like '/url?q=https://en.wikipedia.org/wiki/...'
        if '/url?q=' in href:
            wikipedia_link = href.split('/url?q=')[1].split('&')[0]
            break

if wikipedia_link:
    print("Wikipedia page link:", wikipedia_link)
else:
    print("No Wikipedia link found in the search results.")
```

**Above method might not always work, coz google wont allow all requests**

better use wikipedia api

```python
import requests

def get_wikipedia_link(query):
    # Wikipedia API endpoint for searching
    api_url = "https://en.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "list": "search",
        "srsearch": query,
        "format": "json"
    }
    response = requests.get(api_url, params=params)
    data = response.json()
    search_results = data.get("query", {}).get("search", [])
    if search_results:
        # Get the title of the first search result
        title = search_results[0]["title"]
        # Construct the Wikipedia page URL
        page_url = f"https://en.wikipedia.org/wiki/{title.replace(' ', '_')}"
        return page_url
    else:
        return "No Wikipedia page found for your query."

user_input = input("Enter a topic to search on Wikipedia: ")
link = get_wikipedia_link(user_input)
print("Wikipedia page link:", link)
```

```python
response= requests.get(link)
soup= BeautifulSoup(response.text, 'html.parser')
```

```python
corpus = ''

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'
    
corpus = corpus.strip()

for i in range(500):
    corpus = corpus.replace('['+str(i)+']','')
```
